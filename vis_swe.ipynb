```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
from pdebench.models.fno.utils import FNODatasetSingle
```

```python
dataset = h5py.File("data/SWE/swe_train.h5", "r")
```

# Visualize samples

```python
samples = ["0000", "0800"]
x = dataset["0000"]["grid"]["x"]
y = dataset["0000"]["grid"]["y"]
t_steps = np.arange(0,100,20)
```

```python
fig, axs = plt.subplots(2, 5, figsize=(14, 7))
for s, sample in enumerate(samples):
    for t, time in enumerate(t_steps):
        axs[s, t].pcolormesh(x, y, dataset[sample]["data"][time, :, :, 0])
        axs[s, t].set_title(f"t={time}")
    axs[s, 0].set_ylabel(f"Sample {sample}")
```
