```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import os
```

```python
path = "data/Darcy Flow/"
files = os.listdir(path)
dataset = h5py.File(path + files[0], "r")
```

```python
dataset.keys()
```

# Visualize samples

```python
samples = np.random.randint(0,10000,5)
x = dataset["x-coordinate"]
y = dataset["y-coordinate"]
```

```python
dataset["nu"][0].max()
```

```python
fig, axs = plt.subplots(2, 5, figsize=(14, 7))
for s, sample in enumerate(samples):
    axs[0,s].pcolormesh(x, y, dataset["nu"][sample])
    axs[1,s].pcolormesh(x, y, dataset["tensor"][sample,0])
    axs[0,s].set_title(f"Sample : {sample}")
axs[0, 0].set_ylabel(f"Diffusion coefficient")
axs[1, 0].set_ylabel(f"Darcy flow solution")
```
